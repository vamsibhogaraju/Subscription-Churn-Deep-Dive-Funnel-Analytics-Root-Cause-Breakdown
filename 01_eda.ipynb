# KKBox Revenue Leakage Analysis — EDA
### Data Profiling all the csv's

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')

DATA  = r"./data/raw"
CHUNK = 500_000

def profile(df, name):
    print(f"\n{'='*60}")
    print(f"  {name}")
    print(f"{'='*60}")
    print(f"  Shape         : {df.shape}")
    print(f"  Dtypes:\n{df.dtypes.to_string()}")
    print(f"\n  Null counts:")
    nulls = df.isnull().sum()
    null_pct = (nulls / len(df) * 100).round(2)
    null_df = pd.DataFrame({'null_count': nulls, 'null_%': null_pct})
    print(null_df[null_df['null_count'] > 0].to_string())
    print(f"\n  Duplicated rows: {df.duplicated().sum():,}")
    print(f"\n  Sample (5 rows):")
    display(df.head())
    print(f"\n  Describe (numeric):")
    display(df.describe())
```

## 1. transactions_v2.csv

```python
chunks = []
for chunk in pd.read_csv(f"{DATA}/transactions_v2.csv", chunksize=CHUNK):
    chunks.append(chunk)
txn = pd.concat(chunks, ignore_index=True)

profile(txn, "transactions_v2.csv")

# Extra checks specific to transactions
print("\n  --- Extra Checks ---")
print(f"  Unique msno                : {txn['msno'].nunique():,}")
print(f"  transaction_date range     : {txn['transaction_date'].min()} -> {txn['transaction_date'].max()}")
print(f"  membership_expire_date range: {txn['membership_expire_date'].min()} -> {txn['membership_expire_date'].max()}")
print(f"  actual_amount_paid < 0     : {(txn['actual_amount_paid'] < 0).sum():,}")
print(f"  actual_amount_paid == 0    : {(txn['actual_amount_paid'] == 0).sum():,}")
print(f"  plan_list_price == 0       : {(txn['plan_list_price'] == 0).sum():,}")
print(f"\n  payment_plan_days value counts:")
print(txn['payment_plan_days'].value_counts().to_string())
print(f"\n  payment_method_id value counts:")
print(txn['payment_method_id'].value_counts().to_string())
```

## 2. members_v3.csv

```python
members = pd.read_csv(f"{DATA}/members_v3.csv")

profile(members, "members_v3.csv")

print("\n  --- Extra Checks ---")
print(f"  Unique msno                : {members['msno'].nunique():,}")
print(f"  registration_init_time range: {members['registration_init_time'].min()} -> {members['registration_init_time'].max()}")
print(f"\n  bd (age) distribution:")
print(f"    min={members['bd'].min()} | max={members['bd'].max()} | mean={members['bd'].mean():.1f} | median={members['bd'].median()}")
print(f"    bd < 0       : {(members['bd'] < 0).sum():,}")
print(f"    bd == 0      : {(members['bd'] == 0).sum():,}")
print(f"    bd > 100     : {(members['bd'] > 100).sum():,}")
print(f"    bd 10-80     : {members['bd'].between(10, 80).sum():,}")
print(f"\n  gender value counts:")
print(members['gender'].value_counts(dropna=False).to_string())
print(f"\n  registered_via value counts:")
print(members['registered_via'].value_counts(dropna=False).to_string())
print(f"\n  city value counts (top 10):")
print(members['city'].value_counts(dropna=False).head(10).to_string())
```

## 3. user_logs_v2.csv

```python
# user_logs is large — sample first 2M rows for EDA, then get shape from full scan
sample_chunks = []
total_rows = 0
total_users = set()

for i, chunk in enumerate(pd.read_csv(f"{DATA}/user_logs_v2.csv", chunksize=CHUNK)):
    total_rows += len(chunk)
    total_users.update(chunk['msno'].unique())
    if i < 4:  # keep first 2M rows for profiling
        sample_chunks.append(chunk)

logs_sample = pd.concat(sample_chunks, ignore_index=True)

print(f"  Full file: {total_rows:,} rows | {len(total_users):,} unique users")
print(f"  Sample used for profiling: {len(logs_sample):,} rows")

profile(logs_sample, "user_logs_v2.csv (first 2M rows sample)")

print("\n  --- Extra Checks ---")
print(f"  date range (sample): {logs_sample['date'].min()} -> {logs_sample['date'].max()}")
print(f"  total_secs < 0     : {(logs_sample['total_secs'] < 0).sum():,}")
print(f"  num_unq == 0       : {(logs_sample['num_unq'] == 0).sum():,}")
print(f"  num_100 > num_unq  : {(logs_sample['num_100'] > logs_sample['num_unq']).sum():,}")
print(f"\n  total_secs stats:")
print(logs_sample['total_secs'].describe().to_string())
```

## 4. train_v2.csv

```python
train = pd.read_csv(f"{DATA}/train_v2.csv")

profile(train, "train_v2.csv")

print("\n  --- Extra Checks ---")
print(f"  Unique msno      : {train['msno'].nunique():,}")
print(f"  is_churn counts  :\n{train['is_churn'].value_counts(dropna=False).to_string()}")
print(f"  Churn rate       : {train['is_churn'].mean()*100:.2f}%")
```

## 5. Cross-file Join Quality

```python
txn_users  = set(txn['msno'].unique())
mem_users  = set(members['msno'].unique())
log_users  = total_users
train_users = set(train['msno'].unique())

print("Cross-file msno overlap")
print(f"{'='*50}")
print(f"  txn unique msno       : {len(txn_users):,}")
print(f"  members unique msno   : {len(mem_users):,}")
print(f"  user_logs unique msno : {len(log_users):,}")
print(f"  train unique msno     : {len(train_users):,}")
print()
print(f"  txn -> members match  : {len(txn_users & mem_users):,} / {len(txn_users):,} ({len(txn_users & mem_users)/len(txn_users)*100:.1f}%)")
print(f"  txn -> logs match     : {len(txn_users & log_users):,} / {len(txn_users):,} ({len(txn_users & log_users)/len(txn_users)*100:.1f}%)")
print(f"  txn -> train match    : {len(txn_users & train_users):,} / {len(txn_users):,} ({len(txn_users & train_users)/len(txn_users)*100:.1f}%)")
print(f"  train -> txn match    : {len(train_users & txn_users):,} / {len(train_users):,} ({len(train_users & txn_users)/len(train_users)*100:.1f}%)")
print(f"  train -> members match: {len(train_users & mem_users):,} / {len(train_users):,} ({len(train_users & mem_users)/len(train_users)*100:.1f}%)")
```
